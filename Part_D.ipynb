```python
transactions = []
category_spend = {"food": 0, "travel": 0, "bills": 0}

total_credit = 0
total_debit = 0
transaction_count = 0
highest_transaction = 0
total_amount = 0

print("Enter transactions (type 'done' to stop)\n")

while True:
    user_input = input("Enter amount (or 'done'): ")

    if user_input.lower() == "done":
        break

    try:
        amount = float(user_input)
    except:
        print("Invalid amount. Try again.")
        continue

    t_type = input("Enter type (credit/debit): ").lower()

    if t_type not in ["credit", "debit"]:
        print("Invalid transaction type.")
        continue

    category = input("Enter category (food/travel/bills): ").lower()

    if category not in category_spend:
        print("Invalid category.")
        continue

    # Store transaction
    transactions.append((amount, t_type, category))
    transaction_count += 1
    total_amount += amount

    if amount > highest_transaction:
        highest_transaction = amount

     # Credit / Debit logic
    if t_type == "credit":
        total_credit += amount
    else:
        total_debit += amount
        category_spend[category] += amount

    # High value flag
    if amount > 10000:
        print("⚠ High value transaction detected!")

print("\n--- BAR CHART (Last 10 Transactions) ---")

last_transactions = transactions[-10:]

for txn in last_transactions:
    amount = txn[0]
    stars = int(amount // 1000)

    print(f"₹{amount:.2f} | ", end="")
    for _ in range(stars):
        print("*", end="")
    print()

# Summary Calculations
net_balance = total_credit - total_debit
average_amount = total_amount / transaction_count if transaction_count > 0 else 0

print("\n--- SUMMARY ---")
print("Total Transactions:", transaction_count)
print("Total Credits: ₹", total_credit)
print("Total Debits: ₹", total_debit)
print("Net Balance: ₹", net_balance)
print("Highest Transaction: ₹", highest_transaction)
print("Average Transaction Amount: ₹", round(average_amount, 2))

print("\n--- SPENDING BREAKDOWN ---")
for category, amount in category_spend.items():
    print(f"{category.capitalize()}: ₹{amount}")
```
