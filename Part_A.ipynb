```python
import random
import string

# Constants
MIN_ACCEPTABLE_STRENGTH = 5
SPECIAL_CHARS = "!@#$%^&*"

def analyze_password(password):
    score = 0
    max_score = 7
    feedback = []

    # Track character types
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False

    # Check for consecutive repeats
    has_no_triple_repeat = True

    # Iterate through password with for loop
    for i, char in enumerate(password):
        # Check character types
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in SPECIAL_CHARS:
            has_special = True

        # Check for 3+ consecutive identical characters
        if i >= 2 and password[i] == password[i-1] == password[i-2]:
            has_no_triple_repeat = False

    # Length scoring
    length = len(password)
    if length >= 16:
        score += 3
    elif length >= 12:
        score += 2
    elif length >= 8:
        score += 1
    else:
        feedback.append("too short (need ≥8 chars)")

    # Character type scoring (using for loop to check flags)
    type_checks = [
        (has_upper, "uppercase"),
        (has_lower, "lowercase"),
        (has_digit, "digit"),
        (has_special, f"special char ({SPECIAL_CHARS})")
    ]

    for has_type, type_name in type_checks:
        if has_type:
            score += 1
        else:
            feedback.append(type_name)

    # No triple repeat bonus
    if has_no_triple_repeat:
        score += 1
    else:
        feedback.append("has 3+ repeated chars in a row")

    # Determine rating
    if score <= 2:
        rating = "Weak"
    elif score <= 4:
        rating = "Medium"
    elif score <= 6:
        rating = "Strong"
    else:
        rating = "Very Strong"
    return score, max_score, rating, feedback

def display_analysis(password):
    score, max_score, rating, feedback = analyze_password(password)

    print(f">> Strength: {score}/{max_score} ({rating})")

    if feedback:
        missing = ", ".join(feedback)
        print(f">> Missing/ Issues: {missing}")
    else:
        print(f">> All criteria met!")

    return score

def get_password_input():
    while True:
        password = input("\nEnter password: ")
        score = display_analysis(password)

        if score >= MIN_ACCEPTABLE_STRENGTH:
            print(">> Password accepted!")
            return password
        else:
            print(f">> Try again... (need strength ≥ {MIN_ACCEPTABLE_STRENGTH})")

def generate_password():
    print("\n" + "="*40)
    print("PASSWORD GENERATOR")
    print("="*40)

    while True:
        try:
            length = int(input("Enter desired password length (8-32): "))
            if 8 <= length <= 32:
                break
            print("Please enter a number between 8 and 32.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # Character pool
    char_pool = string.ascii_letters + string.digits + string.punctuation

    # Generate password using for loop
    password_chars = []
    for _ in range(length):
        password_chars.append(random.choice(char_pool))

    generated_password = "".join(password_chars)

    print(f"\nGenerated password: {generated_password}")
    display_analysis(generated_password)

    return generated_password

def main():
    print("="*40)
    print("PASSWORD STRENGTH ANALYZER & GENERATOR")
    print("="*40)

    while True:
        print("\n1. Analyze your password")
        print("2. Generate secure password")
        print("3. Exit")

        choice = input("\nSelect option (1-3): ").strip()

        if choice == "1":
            get_password_input()
        elif choice == "2":
            generate_password()
        elif choice == "3":
            print("\nStay secure! Goodbye.")
            break
        else:
            print("Invalid option. Please choose 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
